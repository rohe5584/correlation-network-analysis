```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def readFile(filepath, name, date_col_name):
    file = pd.read_csv(filepath)
    file[date_col_name] = pd.to_datetime(file[date_col_name])
    print(name)
    print("shape: \n", file.shape)
    print("head: \n", file.head(3))
    print("tail: \n", file.tail(3), '\n')
    return file
```

```python
# read stock data
nasdaq_path = './stock_data/nasdaq_all.csv'
dowjones_path = './stock_data/djia_2012.csv'
russel2000_path = './stock_data/russel2000_all.csv'


nasdaq = readFile(nasdaq_path, 'NASDAQ DATA', 'Date')
dowjones = readFile(dowjones_path, 'DOW JONES DATA', 'Effective date ')
russel2000 = readFile(russel2000_path, 'RUSSEL 2000 DATA', 'Date')
```

```python
# read commodity data
djcommodity_path = './commodity_data/djcommodity_2012.csv'
sp_path = './commodity_data/S&P_2012.csv'

djcommodity = readFile(djcommodity_path, 'DOW JONES COMMODITY DATA', 'Effective date ')
sp = readFile(sp_path, 'S&P COMMODITY DATA', 'Effective date ')
```

```python
tbonds_path = './bond_data/tbonds_2012.csv'
govtcorp_path = './bond_data/govt&corp_2012.csv'


tbonds = readFile(tbonds_path, 'T-BONDS DATA', 'Effective date ')
govtcorp = readFile(govtcorp_path, 'US GOVERNMENT & CORPORATE BOND DATA', 'Effective date ')
```

```python
def plotHistoric(df, title, price_colname, date_colname):
    plt.figure(figsize=(10,10))
    plt.plot(df[date_colname], df[price_colname])
    plt.xlabel("Date")
    plt.ylabel("$ price")
    plt.title(title)
    plt.show()
    

```

```python
plotHistoric(nasdaq, 'NASDAQ HISTORIC DATA', 'Open', 'Date')
```

```python
plotHistoric(dowjones, 'DOW JONES HISTORIC DATA', 'Dow Jones Industrial Average', 'Effective date ')
```

```python
plotHistoric(russel2000, 'RUSSEL 2000 HISTORIC DATA', 'Open', 'Date')
```

```python
plotHistoric(tbonds, 'T-Bonds Historic Data', 'S&P U.S. Treasury Bond Index', 'Effective date ')
```

```python
plotHistoric(govtcorp, 'US GOVERNMENT & CORPORATE BOND DATA', 'S&P U.S. Government & Corporate Bond Index', 'Effective date ')
```

```python
plotHistoric(djcommodity, "DOW JONES COMMODITY HISTORIC DATA", 'Dow Jones Commodity Index', 'Effective date ')
```

```python
plotHistoric(sp, "S&P HISTORIC DATA", "S&P GSCI", "Effective date ")
```

```python
def add_normalize(df, colname):
    df['normalized_price']= (df[colname]-df[colname].mean())/df[colname].std()
```

```python
add_normalize(nasdaq, "Close")
nasdaq.head
```

```python
add_normalize(dowjones, 'Dow Jones Industrial Average')
dowjones.head
```

```python
add_normalize(russel2000, "Close")
russel2000.head
```

```python
# align stock data so they all have the same dates

nasdaq = nasdaq[~(nasdaq['Date'] < '2012-10-31')]
russel2000 = russel2000[~(russel2000['Date'] < '2012-10-31')]
dowjones = dowjones.drop(index=2525)
```

```python
# check that they have the same number of rows
print(dowjones.shape, russel2000.shape, nasdaq.shape)
```

```python
date = nasdaq['Date']
plt.figure(figsize=(10,10))
plt.plot(date, nasdaq['normalized_price'], label = "NASDAQ")
plt.plot(date, dowjones['normalized_price'], label = "DOW JONES")
plt.plot(date, russel2000['normalized_price'], label="RUSSEL 2000")
plt.xlabel("Date")
plt.ylabel("Price (normalized)")
plt.title('COMPARISON OF NORMALIZED PRICES OF STOCK DATASETS')
plt.legend()
plt.show()
```
