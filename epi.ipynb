```python
def readFile(filepath, name, date_col_name):
    file = pd.read_csv(filepath)
    file[date_col_name] = pd.to_datetime(file[date_col_name])
    print(name)
    print("shape: \n", file.shape)
    print("head: \n", file.head(3))
    print("tail: \n", file.tail(3), '\n')
    return file
```
